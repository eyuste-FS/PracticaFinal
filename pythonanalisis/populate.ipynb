#### Funciones auxiliares

```python
def query(q, many: bool, *args):
    try:
        mydb = connection.connect(host="localhost", database = 'practica',user="root", passwd="root", use_pure=True)
        cursor = mydb.cursor(buffered=True)
        if many:
            cursor.executemany(q, *args)
        else:
            res = cursor.execute(q, *args)
        mydb.commit()
    except Exception as e:
        res = 'Excepcion ' + str(e)
    finally:
        # cursor.close()
        mydb.close()
    return res
```

```python
def insert(table, campos, data, many: bool = False):
    if not len(campos) or not len(data):
        return 'Empty row'

    q = f'insert into {table} ({", ".join(campos)}) values ({", ".join(["%s"] * len(data[0] if many else data))})'
    res = None
    try:
        mydb = connection.connect(host="localhost", database = 'practica',user="root", passwd="root", use_pure=True)
        cursor = mydb.cursor(buffered=True)
        if many:
            res = cursor.executemany(q, data)
        else:
            res = cursor.execute(q, data)
        mydb.commit()
    except Exception as e:
        res = 'Excepcion ' + str(e)
    finally:
        # cursor.close()
        mydb.close()
    return res
```
